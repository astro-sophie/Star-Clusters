```python
import pandas as pd
from astroquery.vizier import Vizier
Vizier.ROW_LIMIT = 40000 # makes sure you don't lose columns--make sure the number of rows isn't equal to this or you might have to increase it 
```

```python
# query Spitzer (code = II/368) centered on NGC 1977 with a radius of 20 arcminutes
result = Vizier.query_region("NGC 1977", radius="0d20m0s", catalog='II/368') 
result
```
